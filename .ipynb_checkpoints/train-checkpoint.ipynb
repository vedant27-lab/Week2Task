```python
#!pip install numpy
#!pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
X = np.linspace(-2, 2, 1000).reshape(-1, 1)
y = X**3 + 0.3 * np.random.randn(*X.shape)
```

```python
plt.scatter(X, y, s=5)
plt.title("Noisy Cubic Data")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
from model import Layer_Dense, Loss_MSE, ReLU, Sigmoid, SGD
```

```python
layer1 = Layer_Dense(1, 64)
activation1 = ReLU()
```

```python
layer2 = Layer_Dense(64, 64)
activation2 = ReLU()
```

```python
layer3 = Layer_Dense(64, 1)
```

```python
loss_function = Loss_MSE()
optimizer = SGD(learning_rate=0.01)
```

```python
epochs = 500
batch_size = 32
losses = list()
```

```python
for epoch in range(epochs):
    indices = np.arange(X.shape[0])
    np.random.shuffle(indices)
    X_shuffled = X[indices]
    y_shuffled = y[indices]
    batch_loss = []

    for i in range(0, X.shape[0], batch_size):
        X_batch = X_shuffled[i:i+batch_size]
        y_batch = y_shuffled[i:i+batch_size]

        layer1.forward(X_batch)
        activation1.forward(layer1.output)

        layer2.forward(activation1.output)
        activation2.forward(layer2.output)

        layer3.forward(activation2.output)
        y_pred = layer3.output

        loss = loss_function.forward(y_pred, y_batch)
        batch_loss.append(loss)

        loss_function.backward(y_pred, y_batch)
        layer3.backward(loss_function.dinputs)

        activation2.backward(layer3.dinputs)
        layer2.backward(activation2.dinputs)

        activation1.backward(layer2.dinputs)
        layer1.backward(activation1.dinputs)


        optimizer.update(layer1)
        optimizer.update(layer2)
        optimizer.update(layer3)

        mean_loss = np.mean(batch_loss)
        losses.append(mean_loss)

        if epoch % 50 == 0 or epoch == epochs - 1:
            print(f"Epoch {epoch+1}/{epochs}, Loss: {mean_loss:.4f}")
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.title("Training Loss Curve")
plt.grid(True)
plt.show()
```

```python
layer1.forward(X)
activation1.forward(layer1.output)

layer2.forward(activation1.output)
activation2.forward(layer2.output)

layer3.forward(activation2.output)
predictions = layer3.output
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y, label="True Data", s=5)
plt.plot(X, predictions, color='r', label="NN Predictions", linewidth=2)
plt.title("Model Fit")
plt.legend()
plt.grid(True)
plt.show()
```
